```python
import numpy as np
import matplotlib.pyplot as plt
from gensim.models import Word2Vec
from sklearn.decomposition import PCA
words = ['king', 'queen', 'man', 'woman', 'paris', 'france', 'london', 'england']
corpus = [
    ['king', 'queen', 'man', 'woman'],
    ['paris', 'france', 'london', 'england'],
    ['king', 'man', 'queen', 'woman'],
    ['london', 'paris', 'france', 'england']
]
model = Word2Vec(corpus, vector_size=100, window=2, min_count=1, workers=4)
word_vectors = np.array([model.wv[word] for word in words])
pca = PCA(n_components=2)
reduced_vectors = pca.fit_transform(word_vectors)

plt.figure(figsize=(10, 6))
for i, word in enumerate(words):
    plt.scatter(reduced_vectors[i, 0], reduced_vectors[i, 1])
    plt.annotate(word, (reduced_vectors[i, 0], reduced_vectors[i, 1]), textcoords="offset points", xytext=(0, 5), ha='center')

plt.title('Word Embeddings Visualized with PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.grid()
plt.show()
```
